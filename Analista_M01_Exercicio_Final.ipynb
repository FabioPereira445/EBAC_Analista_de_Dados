```python
#PARTE 2 OBJETIVO:Extraia da string abaixo o valor da taxa selic na vari치vel selic e o valor do ano na vari치vel ano . Imprima os valores na tela.

noticia = 'Selic vai a 2,75% e supera expectativas; 칠 a primeira alta em 6 anos.'

# Extraindo o valor da taxa Selic
selic_str = noticia.split()[3].replace(',', '.').replace('%', '')
selic = float(selic_str)

# Extraindo o ano da not칤cia
ano = int(noticia.split()[-2])

# Imprimindo os valores na tela
print(f"Valor da Selic: {selic:.2f}")
print(f"Ano: {ano}")


```
