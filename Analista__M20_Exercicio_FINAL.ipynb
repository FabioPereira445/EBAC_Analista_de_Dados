### **1.1. Nativo**

Neste primeira atividade, você deve ler o código abaixo implementado em Python nativo e responder a seguinte pergunta:

> O que o código abaixo computa?

**Resposta**: Os dias solicitados e os incidentes totais dos mesmos.

```
# Leitura e carregamento do arquivo no ambiente

data = None #criada uma lista vazia nomeada como data


with open(file='traffic.csv', mode='r', encoding='utf8') as fp:

#código  Ignora a primeira linha (cabeçalho)
  fp.readline()

   #código  Ler todo o conteúdo do arquivo em uma string
  data = fp.read()

# 2. Análise e Contagem de Acidentes: analytics

day = 14            # Dia inicial da análise
incidents = 0       # Contador de acidentes

# Dicionário para armazenar contagem de acidentes por dia
incident_by_day = dict()

for timebox in data.split(sep='\n'):

  timebox_data = timebox.split(sep=';')

  # --
  # -- inicio da computação escalar
  # --

  for incident in timebox_data[1: len(timebox_data)-1]:
    incidents = incidents + int(incident)

  # --
  # -- fim da computação escalar
  # --

  try:

    half_hour = int(timebox_data[0])

    if half_hour == 27:
      incident_by_day[day] = incidents
      day = day + 1
      incidents = 0

  except ValueError:
    continue

# --  # Apresentar resultados

for day in incident_by_day:

  print(f'{day}: {incident_by_day[day]}')
```

### **1.2. NumPy**

Substitua o trecho do código do algoritmo que utiliza da **computação escalar** por um que utiliza da **computação vetorial**. Use o pacote NumPy.

```
# -- read

#código  Abre o arquivo CSV no modo de leitura e codificação UTF-8
with open(file='traffic.csv', mode='r', encoding='utf8') as fp:

    #código  Ignora a primeira linha do arquivo (cabeçalho)
    fp.readline()

    #código  Ler todo o conteúdo do arquivo em uma string e armazena na variável 'data'
    data = fp.read()

# -- analytics

#código  Importa a biblioteca NumPy para operações vetoriais
import numpy as np

#código  Define a variável 'day' para armazenar o dia inicial da análise
day = 14

#código  Define a variável 'incidents' para contabilizar o número de acidentes
incidents = 0

#código  Cria um dicionário 'incident_by_day' para armazenar a contagem de acidentes por dia
incident_by_day = dict()

#código  Itera sobre cada linha do arquivo CSV (separada por '\n')
for timebox in data.split(sep='\n'):

    #código  Separa os dados de cada linha por ponto e vírgula (';')
    timebox_data = timebox.split(sep=';')

    # --
    # -- inicio da computação vetorial
    # --

  #código  Converte a variável 'incidents' em um array NumPy
    incidents_array = np.array(incidents)

    #código  Converte os dados da linha atual (exceto a primeira e a última linha) em um array NumPy
    timebox_array = np.array(timebox_data[1: len(timebox_data)-1])

    #código  Itera sobre o array 'timebox_array' e soma os valores ao contador de acidentes
    for incidents_array in timebox_array:
        incidents = incidents + int(incidents_array)

    # --
    # -- fim da computação vetorial
    # --

    #código  Tenta converter a meia-hora da linha atual para um inteiro
    try:
        half_hour = int(timebox_data[0])

        #código  Verifica se é a última meia-hora do dia (27)
        if half_hour == 27:
            #código  Armazena a contagem de acidentes do dia atual no dicionário
            incident_by_day[day] = incidents

            #código  Incrementa o dia para a próxima análise
            day = day + 1

            #código  Reinicializa o contador de acidentes para o próximo dia
            incidents = 0

    #código  Ignora linhas com valores inválidos na meia-hora
    except ValueError:
        continue

# -- results

#código Imprime a contagem de acidentes por dia
for day in incident_by_day:
    print(f'Dia {day}: {incident_by_day[day]} acidentes')
```
