## 1\. Preços `outliers` de diamante

```
#importando biblioteca seaborn
import seaborn as sns

data = sns.load_dataset("diamonds")  # Carrega o dataset 'diamonds' da biblioteca Seaborn
data.head()  # Exibe as primeiras linhas do dataset para visualização inicial




```

```
# Criação do Boxplot com estilo 'whitegrid'
with sns.axes_style("darkgrid"):
    diamonds = sns.boxplot(x=data["price"])  # Gera o boxplot utilizando a coluna 'price'
    diamonds.set(title="Preços outliers diamantes", xlabel="Preço")  # Define título e rótulo do eixo X
```

**Insight do gráfico 1**:
* Os preços dos diamantes variam de 0 a 11.500 aproximadamente, com a maioria dos valores se concentrando entre 0 e 5.500. Essa faixa representa a maior parte dos diamantes disponíveis no mercado, geralmente com características mais comuns.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
 # Carregamento do dataset 'diamonds' da biblioteca Seaborn
import seaborn as sns

data = sns.load_dataset("diamonds")

 # Exibe as primeiras linhas do dataset para visualização inicial
data.head()
```

```
# gráfico do exercício número 2
# Seleção das colunas relevantes
diamonds = data[["price", "carat", "clarity"]]
diamonds.head()  # Exibe as primeiras linhas do subset para verificação
```

```
# Criação do Scatterplot com estilo 'ticks' e HUE para 'clarity'
with sns.axes_style("ticks"):
    grafico = sns.scatterplot(
        data=diamonds,
        x="price",
        y="carat",
        hue="clarity",
        palette="pastel",
    )
    grafico.set(
        title="Correlação entre Preço e Peso por Clareza",
        xlabel="Preço",
        ylabel="Peso",
    )
    grafico.figure.set_size_inches(w=50 / 2.54, h=20 / 2.54)  # Ajuste do tamanho da figura
    grafico.get_legend().set_title("Clareza")  # Define título para a legenda
```

**Insight do gráfico 2**:
 * A escolha da clareza do diamante pode ser crucial para obter o melhor custo-benefício, considerando o peso desejado.
Diamantes com menor clareza (I1) podem ser opções mais acessíveis para quem busca pedras maiores, mas é importante considerar outros fatores como lapidação e cor.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
import seaborn as sns

# Carregamento do dataset 'mpg' da biblioteca Seaborn
data = sns.load_dataset("mpg")
data.head()  # Exibe as primeiras linhas do dataset para visualização inicial

# Seleção das colunas relevantes
cars = data[["weight", "origin"]]
cars.head()  # Exibe as primeiras linhas do subset para verificação

# Criação do Histograma com estilo 'white' e HUE para 'origin'
with sns.axes_style("white"):
    grafico = sns.histplot(
        data=cars,
        x="weight",
        hue="origin",
           palette="dark",
        kde=True,
    )
    grafico.set(
        title="Distribuição Peso dos Carros por Origem",
        xlabel="Peso",
        ylabel="Contagem",
    )
    grafico.get_legend().set_title("Origem")  # Define título para a legenda
```

**Insight do gráfico 3**:

**Faixa de 1.500 a 2.000 kg:**

Japoneses e Europeus: Dominam a cena, com maior probabilidade de encontrarmos carros dessas origens nessa faixa de peso.

**Faixa de 2.500 a 5.000 kg:**

Declínio Europeu e Japonês: Observe a queda na presença de carros europeus e japoneses nessa faixa. Eles praticamente somem, dando lugar aos americanos.

Americanos em Ascensão: É nesse momento que os carros americanos assumem o protagonismo, se tornando a principal força nessa categoria de peso mais elevado.

**Conclusão:**

Peso e Origem: Uma relação intrigante! Quanto maior o peso do carro, maior a probabilidade de ser americano. Já os carros japoneses e europeus concentram-se na faixa de 1.500 a 2.250 kg, perdendo espaço para os americanos à medida que o peso aumenta.
