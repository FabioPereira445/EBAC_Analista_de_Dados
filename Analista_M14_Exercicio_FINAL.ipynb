## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
import seaborn as sns

# Carrega o conjunto de dados "diamonds"
data = sns.load_dataset("diamonds")

# Cria um gráfico de barras para visualizar o preço médio por tipo de corte
diamonds = sns.barplot(data=data, x="cut", y="price", ci=None, palette="dark")

# Define o título, rótulos dos eixos e legenda
diamonds.set(title='Preço Médio do Diamante por Tipo de Corte', xlabel='Tipo de Corte', ylabel='Preço Médio')
```

**Insight do gráfico 1**: O corte premium apresenta o preço mais elevado, o corte fair vem em segundo lugar,já o corte very good aparece com o terceiro maior preço, o corte good em quarto e o ideal em quinto e último lugar.Essas informações podem estar ligadas e relacionadas a oferta e procura ao diamante sendo trabalhado das respectivas formas exigidos pelos clientes afetando diretamente o preço e disponibilidade.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data_flights = sns.load_dataset("flights")
data_flights.head()
```

```

# gráfico do exercício 2
import numpy as np

flights = data_flights.query('month == "Dec"')

with sns.axes_style('whitegrid'):
  grafic_flights = sns.lineplot(x='year', y='passengers', data=flights);
  grafic_flights.set(title = 'Número de passageiros em dezembro por ano',
                      xlabel = 'Ano',
                      ylabel = 'Passageiros',
                      ylim = [100, 500]);


```

**Insight do gráfico 2**: Ao analisar o gráfico de linha chegamos a conclusão que o número de passageiros no mês de dezembro de cada ano teve aumento de forma constante a cada 2 anos,somente entre 1956 e 1958 teve uma estagnação durando por volta de 1 ano entre 1957 e 1958 voltando após esse período, ao seu padrão de crescimento conquistado ao longo dos anos.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

flights_A = sns.load_dataset("flights")
flights_A.head()
```

```

# gráfico do exercício 3

flights_mes = flights_A.query('year >= 1949 and year <= 1959')

with sns.axes_style('whitegrid'):
  grafic_flights_mes = sns.lineplot(x='month', y='passengers', hue='year', data=flights_mes);
  grafic_flights_mes.set(title = 'Número de passageiros por mês',
                      xlabel = 'Meses',
                      ylabel = 'Passageiros')
  grafic_flights_mes.grid(alpha=0.5)
  grafic_flights_mes.legend(title='Ano',loc=[1.01,0])

```

**Insight do gráfico 3**:Ao analisar o comportamento do gráfico podemos chegar a algumas conclusões referente ao comportamento do consumidor em alguns meses.Percebemos que houve aumento de viagens/passageiros a partir do mês de maio chegando ao ápice em agosto em praticamente  todos os anos.Entre março,abril e maio manteve uma certa constância com leves quedas no número de passageiros. Podendo tomar decisões com base nesses dados em várias áreas da companhia como no marketing,no preço das passagens,na contratação temporária de mais funcionários no período de pico para manter a qualidade de atendimento,entre outros.
