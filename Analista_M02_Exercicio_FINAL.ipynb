1. Listas

Criei uma lista chamada filmes com o nome dos 10 primeiros filmes mais bem avaliados no site no IMDB. Imprima o resultado.

```python

filmes = ['F1-Um Sonho de Liberdade', 'F2-O Poderoso Chefão I', 'F3-Batman: O Cavaleiro das Trevas', 'F4-O Poderoso Chefão II',
          'F5-12 Homens e uma Sentença', 'F6-A Lista de Schindler', 'F7-O Senhor dos Anéis: O Retorno do Rei',
          'F8-Pulp Fiction: Tempo de Violência', 'F9-O Senhor dos Anéis: A Sociedade do Anel', 'F10-Três Homens em Conflito']
```

```python
print(filmes)
```

2-Simule a movimentação do ranking. Utilize os métodos insert e pop para trocar a posição do primeiro e do segundo filme da lista. Imprima o resultado.

```python

# Para Remover o primeiro filme da lista
primeiro_filme = filmes.pop(0)

# Inserir o primeiro filme na segunda posição
filmes.insert(1, primeiro_filme)

# # Imprimir a lista atualizada
print(filmes)
```

2°.Exercício **Conjuntos**

Parte 1:Imagine que Aconteceu um erro no seu ranking. Simule a duplicação dos três últimos filmes da lista. Imprima o resultado.

```python

# Duplicando os últimos três filmes usando a técnica de slicing
filmes_duplicados = filmes[-3:]

# Adicionando os filmes duplicados à lista original
filmes.extend(filmes_duplicados)

# Imprimindo a lista atualizada
print(filmes)
```

Parte 2: Utilizando a conversão set e list para remover os valores duplicados. Imprimindo o resultado.

```python
print(filmes)#lista com os filmes duplicados
```

```python

# Convertendo a lista list em um conjunto  set para remover duplicatas
filmes_set = set(filmes)

# Convertendo o conjunto set de volta para uma lista list
filmes = list(filmes_set)

# Imprimindo a lista sem duplicatas
print(filmes)
```
