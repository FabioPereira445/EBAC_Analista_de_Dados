# **Exercícios**

## 1\. Pinguins

Neste exercício, vamos utilizar uma base de dados com informações sobre penguins. A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```
 #Baixando bibliotecas necessárias
import numpy as np
import pandas as pd
import seaborn as sns
```

```
data = sns.load_dataset('penguins')
```

```
data.head()
```

### **1.1. Valores nulos**

A base de dados possui valores faltantes, utilize os conceitos da aula para trata-los.

```
#Comando analisa informaçoes do dataframe
print(data.info())


#Comando  soma os dados faltantes presentes no dataframe
print(data.isna().sum())
```

```
#comando  remove linhas do dataframe com valores nulos
data = data.dropna(how="any", axis=0)
data.head()
```

### **1.2. Variáveis numéricas**

Identifique as variáveis numéricas e crie uma nova coluna **padronizando** seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_std*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.2

data[["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]].head()
```

```
#Comando realiza padronizacao da coluna "bill_length_mm"

#  Comando Encontra media
media_length = data["bill_length_mm"].mean()

#  Encontra desvio padrao
desvio_padrao_length = data['bill_length_mm'].std()


data["bill_length_mm_std"] = data["bill_length_mm"].apply(lambda nota: (nota - media_length) / desvio_padrao_length )
data.head()
```

```
# Comando realiza padronização da coluna "bill_depth_mm"

# Comando calcula media
media_depth = data["bill_depth_mm"].mean()

# Comando calcula desvio padrao
desvio_padrao_depth = data["bill_depth_mm"].std()

data["bill_depth_mm_std"] = data["bill_depth_mm"].apply(lambda nota: (nota - media_depth) / desvio_padrao_depth)
data.head()
```

```
# Comando calcula media
media_flipper_length = data["flipper_length_mm"].mean()

#Comando calcula desvio padrao
desvio_padrao_flipper_length = data["flipper_length_mm"].std()

data["flipper_length_mm_std"] = data["flipper_length_mm"].apply(lambda nota: (nota - media_flipper_length) / desvio_padrao_flipper_length)
data.head()
```

```
# Comando calcula media
media_body_mass = data["body_mass_g"].mean()

#Comando calcula desvio padrao
desvio_padrao_body_mass = data["body_mass_g"].std()

data["body_mass_g_std"] = data["body_mass_g"].apply(lambda nota: (nota - media_body_mass) / desvio_padrao_body_mass)
data.head()
```

### **1.3. Variáveis categóricas**

Identifique as variáveis categóricas nominais e ordinais, crie uma nova coluna aplicando a técnica correta de conversão a seus valores. A nova coluna deve ter o mesmo nome da coluna original acrescidade de "*_nom*" ou "*_ord*".

> **Nota**: Você não deve tratar a variável resposta.

```
# resposta da questão 1.3

# Encontrando atributo categorico nominal
data[['sex']].head()

# Comando aplica tecnica de codificacao chamada 'one hot encoding'.
data['sex_m_nom'] = data['sex'].apply(lambda sex: 1 if sex == 'Male' else 0)
data['sex_f_nom'] = data['sex'].apply(lambda sex: 1 if sex == 'Female' else 0)

data.head()
```

```
#  Comadno aplica atributo categorico ordinal
data[["island"]].head()

data['island'].drop_duplicates()

#comando classifica as ilhas presente no atributo "island"
ilhas_pinguim = {
    "Torgersen": 3,
    "Biscoe": 2,
    "Dream": 1,
}

data["island_ord"] = data["island"].apply(lambda level: ilhas_pinguim[level])

data.head()
```

Descarte as colunas originais e mantenha apenas a variável resposta e as variáveis preditivas com o sufixo *_std*", *_nom*" e "*_ord*".
