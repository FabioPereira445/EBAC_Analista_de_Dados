## 3\. Strings

Aplique três métodos distintos na *string* abaixo, você pode conferir alguns métodos neste [link](https://www.w3schools.com/python/python_ref_string.asp):

```
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'
# 1. Maiúsculas com método .upper
cancao_maiuscula = cancao.upper()
print(f"String em maiúsculas: {cancao_maiuscula}")

# 2. Minúsculas com métofo .lower
cancao_minuscula = cancao.lower()
print(f"String em minúsculas: {cancao_minuscula}")

# 3. Remover espaços com método .replace
cancao_sem_espaco = cancao.replace(" ", "")
print(f"String sem espaços: {cancao_sem_espaco}")
```

Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
#METODO 1
#PARTE 2 OBJETIVO:Extraia da string abaixo o valor da taxa selic na variável selic e o valor do ano na variável ano . Imprima os valores na tela.

noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

# Extraindo o valor da taxa Selic
selic_str = noticia.split()[3].replace(',', '.').replace('%', '')
selic = float(selic_str)

# Extraindo o ano da notícia
ano = int(noticia.split()[-2])

# Imprimindo os valores na tela
print(f"Valor da Selic: {selic:.2f}")
print(f"Ano: {ano}")

```

```
#METODO 2 MAIS EFICIENTE
import re as regex

noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'

resultados = regex.search(r"Selic vai a (\d+\,\d+)\% e supera expectativas; é a primeira alta em (\d+) anos.", noticia)

if resultados:
    selic = resultados.group(1)# Extrai valor da selic do grupo 1
    ano = "Ano não encontrado na expressão regular"  # Valor padrão se o ano não for encontrado
    print(f"Selic: {selic}")

    # **Opcional:** Se desejar imprimir o ano extraído, utilize o grupo 2 da expressão regular.

if resultados:
    selic = resultados.group(2) # Extrai valor da selic do grupo 2 (correto)
    ano = "Ano não encontrado na expressão regular"  # Valor padrão se o ano não for encontrado
    print(f"Selic: {selic}")
else:
    print("Informação não encontrada.")
```

Pistas:

Compreendendo a tabela:

A tabela da verdade mostra o resultado de operações booleanas para todas as combinações de valores (Verdadeiro e Falso) dos operandos.
Na primeira coluna, você encontra as combinações de valores dos operandos.
Nas colunas seguintes, você encontra os resultados das operações (AND, OR, NOT).

Na célula de código, você tem as variáveis a e b com seus valores definidos:

Aplicando a tabela:

Na tabela da verdade, encontre a linha onde a é Falso e b é Verdadeiro.
O valor na coluna not a & b nessa linha será o valor da variável x.
Calculando o valor de x:

Na célula de código, utilize a operação not a & b para calcular o valor de x.
Imprimindo o resultado:

Utilize a função print() para exibir o valor de x na tela.

Utilize a tabela da verdade para responder: qual o valor da variável x?

```
a = False
b = True

x = not a & b
```

```
# Tabela da Verdade
#| a | b | not a | not a & b | a & b | a or b |
#|---|---|---|---|---|---|
#| F | F | T | F | F | F |
#| F | T | T | T | F | T |
#| T | F | F | F | F | F |
#| T | T | F | F | T | T |

# Calculando o valor de x
x = not a & b
print(f"Valor de x: {x}")
```
